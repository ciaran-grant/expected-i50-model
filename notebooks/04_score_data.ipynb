```python
import pandas as pd
import numpy as np
import joblib
import warnings
warnings.filterwarnings('ignore')

from expected_i50_model.config import modelling_file_path
from expected_i50_model.src.modelling_data_contract import ModellingDataContract

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999

%load_ext autoreload
%autoreload 2
```

New Model

```python
new_version = 1
model_name = "expected_i50"
model_file_path = "/Users/ciaran/Documents/Projects/AFL/git-repositories/expected-i50-model/model_outputs/models"
preproc_file_path = "/Users/ciaran/Documents/Projects/AFL/git-repositories/expected-i50-model/model_outputs/preprocessors"
```

Response

```python
RESPONSE = ModellingDataContract.RESPONSE
```

Features

```python
FEATURES = ModellingDataContract.feature_list
```

Load Data

```python
modelling_data = pd.read_csv(modelling_file_path)
modelling_data.tail()
```

```python
X, y = modelling_data.drop(columns=[RESPONSE]), modelling_data[RESPONSE]
```

```python
X_preproc = X[FEATURES]
```

Load Model

```python
super_xgb = joblib.load(model_file_path + "/" + model_name + "_v" + str(new_version) + ".joblib")
```

```python
super_xgb.xgb_model
```

Get Predictions

```python
preds = super_xgb.predict(X_preproc)
```

```python
probas = super_xgb.predict_proba(X_preproc)[:, 1]
```

```python
cal_probas = super_xgb.predict_proba(X_preproc, calibrate=True)
```

Export Data

```python
data = modelling_data.reset_index(drop = True)
data_export = pd.concat([data, X_preproc.reset_index(drop=True)], axis=1)
data_export['xInside50_preds'] = preds
data_export['xInside50_probas'] = probas
data_export['xInside50_cal_probas'] = cal_probas
data_export.to_csv("/Users/ciaran/Documents/Projects/AFL/git-repositories/expected-i50-model/data/predictions/expected_inside50.csv", index = False)
data_export.head()
```
